```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.linalg import eigh
from sklearn.datasets  import load_digits
```

```python
mnist = load_digits()
data = mnist['images']
data.shape
```

```python
plt.imshow(data[0], cmap="gray")
plt.show()
```

```python
data = data.reshape(data.shape[0], -1)
data.shape
```

```python
mean_vec = np.mean(data, axis=0)
```

```python
data_norm = data - mean_vec
covariance_matrix = (data_norm).T.dot((data_norm)) / (data_norm.shape[0])
```

```python
eigen_vals, eigen_vects = eigh(covariance_matrix)
```

```python
eigen_vals
```

# Task

## 1. Sort the "eigen_values and eigen_vectors" according to the decreasing order of the Eigen value

```python
# Code
idx=eigen_vals.argsort()[::-1]
eigen_vals=eigen_vals[idx]
print(idx)
```

```python
eigen_vects = eigen_vects[:,idx]
eigen_vects.shape
```

```python
num_of_component = 2
```

## 2. Take the first num_of_component components from the eigen_values and eigen_vectors

```python
# Code
eigen_values = eigen_vals[:num_of_component]
```

```python
eigen_values
```

```python
eigen_vectors = eigen_vects[:,:num_of_component]
```

```python
eigen_vectors.shape
```

## 3. Perform the dimensionality reduction to get lower dimension data by multiplying data matrix with first num_of_component components eigen_vectors matrix

```python
result = np.dot(data,eigen_vectors)
```

```python
result.shape
```

```python
df = pd.DataFrame(data = result, columns = ("1st principal","2nd principal"))
```

```python
df.head()
```

## 5. Find the num_of_component according to the variance explained of the eigen value
.
* Calculate the variance explained by each eigen value as  $$ explained\_variance[i] =  \frac{ eigen\_values[i]}{\sum_{j=1}^{N} eigen\_values[j]} $$
* Calculate the total variance explained by the first K eigen value using:  $$ total\_variance[K] = \sum_{i=1}^{K} explained\_variance[i] $$
* N = Number of eigen value i.e 64 and K is number of eigen values that varies from 1 to N
* For graph plotting
  * Do line plot where K value in X-axis, total_variance[K] value in Y-axis  
    using parameter marker='o', linestyle='--', color='b'
  * Put Proper x-label, y-label and title in the graph
  * Show vertical grid line in the graph
  * Plot one horizontal red color line at the choosen threshold level. 
  * Write down the conclusion that you get from the graph




```python
explained_variance = [np.abs(i)/np.sum(eigen_vals) for i in eigen_vals]
```

```python
total_variance = np.cumsum(explained_variance)
total_variance
```

```python
plt.figure(figsize=(25,10))
plt.plot(total_variance,marker='o',linestyle='--',color='b')
plt.xticks(np.arange(0,64,step=1))
plt.xlabel('Number of Components')
plt.ylabel('Variance(%)')
plt.title('Mnist Dataset Explained Variance')
plt.axhline(y=0.9,color = 'r',linestyle='-')
plt.grid(axis='x')
plt.show
```

If we take 20 features it can represent our dataset with 90% accuracy. If we take 40 features out of 64, it can represent whole our dataset with almost 99% accuracy.
